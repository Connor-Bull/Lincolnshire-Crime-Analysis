```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Lincolnshire_Crime_Data_2021.csv')
df.info()
df.head()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(df.isnull(), cbar = False, cmap = 'viridis')
plt.savefig("null_values.png")
```

```python
df = df.drop(columns=['Crime ID', 'Context', 'Reported by', 'Falls within'], axis = 1)
df.head()
df.tail()
```

```python
df = df.dropna()
plt.figure(figsize=(15,7))
sns.heatmap(df.isnull(), cbar = False, cmap = 'viridis')
plt.savefig("no_null_values.png")
```

```python
pd.value_counts(df['Crime type'])[:10]
```

```python
plt.figure(figsize=(25,10))
plt.title('Amount of Crimes by Crime Type')
plt.ylabel('Crime Type')
plt.xlabel('Amount of Crimes')

df.groupby([df['Crime type']]).size().sort_values(ascending=True).plot(kind='barh')
plt.savefig("Crime_Overview.png")
```

```python
plt.figure(figsize=(30,10))
plt.title('Crime hotspots')
sns.countplot(y= 'Location', data = df, order = df['Location'].value_counts().iloc[:10].index)
plt.savefig("Crime_Location_Overview.png")
```

```python
df_VS = df[df['Crime type'] == 'Violence and sexual offences']
plt.figure(figsize=(30,10))
plt.title('Violence and Sexual offences outcomes')
sns.countplot(y = df_VS['Last outcome category'])
plt.savefig("VS_Outcomes.png")
```

```python
df_CD = df[df['Crime type'] == 'Criminal damage and arson']

plt.figure(figsize=(30,10))


plt.title('Criminal damage and arson outcome')
sns.countplot(y = df_CD['Last outcome category'])
plt.savefig("CD_Outcomes.png")
```

```python
df_PO = df[df['Crime type'] == 'Public order']


plt.figure(figsize=(30,10))


plt.title('Public order outcomes')

sns.countplot(y = df_VS['Last outcome category'])
plt.savefig("PO_Outcomes.png")
```

```python
%%time
df_VS['Date'] = pd.to_datetime(df_VS['Month'])
df_VS['Month'] = df_VS['Date'].apply(lambda x : x.month)
VS_in_months = pd.DataFrame({"VS" : df_VS['Month'].value_counts(),  "month" : df_VS["Month"].value_counts().index}, index = range(12))

VS_in_months.fillna(0, inplace=True)
VS_in_months = VS_in_months.sort_values(['month'], ascending=[1])


```

```python
plt.figure(figsize=(15,7))
plt.plot(VS_in_months['month'],VS_in_months['VS'], label = 'Total In Month')
plt.plot(VS_in_months['month'],VS_in_months['VS'].rolling(window = 2).mean(),color='red', 
         linewidth=5, label='2-months Moving Average' )

plt.title('Violence and Sexual offences per month (2021)', fontsize=16)
plt.xlabel('Months')
plt.legend(prop={'size':16})
plt.tick_params(labelsize=16);


plt.savefig("VS_year.png")
```

```python
%%time
df_CD['Date'] = pd.to_datetime(df_CD['Month'])
df_CD['Month'] = df_CD['Date'].apply(lambda x : x.month)
CD_in_months = pd.DataFrame({"CD" : df_CD['Month'].value_counts(), 
                             "month" : df_CD["Month"].value_counts().index}, index = range(12))
CD_in_months.fillna(0, inplace=True)
CD_in_months = CD_in_months.sort_values(['month'], ascending=[1])

plt.figure(figsize=(15,7))
plt.plot(CD_in_months['month'],CD_in_months['CD'], label = 'Total In Month')
plt.plot(CD_in_months['month'],CD_in_months['CD'].rolling(window = 2).mean(),color='red', linewidth=5, label='2-months Moving Average' )

plt.title('Criminal Damage and Arsen per month (2021)', fontsize=16)
plt.xlabel('Months')
plt.legend(prop={'size':16})
plt.tick_params(labelsize=16);

plt.savefig("CD_year.png")
```

```python
%%time
df_PO['Date'] = pd.to_datetime(df_PO['Month'])
df_PO['Month'] = df_PO['Date'].apply(lambda x : x.month)
PO_in_months = pd.DataFrame({"PO" : df_PO['Month'].value_counts(), 
                             "month" : df_PO["Month"].value_counts().index}, index = range(12))
PO_in_months.fillna(0, inplace=True)
PO_in_months = PO_in_months.sort_values(['month'], ascending=[1])

plt.figure(figsize=(15,7))
plt.plot(PO_in_months['month'],PO_in_months['PO'], label = 'Total In Month')
plt.plot(PO_in_months['month'],PO_in_months['PO'].rolling(window = 2).mean(),color='red', linewidth=5, label='2-months Moving Average' )

plt.title('Public order offences per month (2021)', fontsize=16)
plt.xlabel('Months')
plt.legend(prop={'size':16})
plt.tick_params(labelsize=16);

plt.savefig("PO_year.png")
```

```python
%%time
df['Date'] = pd.to_datetime(df['Month'])
df['Month'] = df['Date'].apply(lambda x : x.month)
in_months = pd.DataFrame({"crime" : df['Month'].value_counts(), 
                             "month" : df["Month"].value_counts().index}, index = range(12))
in_months.fillna(0, inplace=True)
in_months = in_months.sort_values(['month'], ascending=[1])

plt.figure(figsize=(15,7))
plt.plot(in_months['month'],in_months['crime'].rolling(window = 2).mean(),color='red', linewidth=5, label='2-months Moving Average' )

plt.title('Crimes per month (2021)', fontsize=16)
plt.xlabel('Months')
plt.legend(prop={'size':16})
plt.tick_params(labelsize=16);
plt.savefig("Crime_per_month.png")
```
