```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

import matplotlib
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score

# ML Libraries
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

# Evaluation Metrics
from yellowbrick.classifier import ClassificationReport
from sklearn import metrics
```

```python
df = pd.read_csv('Lincolnshire_Crime_Data_2021.csv')
df.info()
df.head()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.isnull(), cbar = False, cmap = 'viridis')
```

```python
df = df.drop(columns=['Crime ID', 'Context', 'Reported by', 'Falls within'], axis = 1)
df.head()
df.tail()
```

```python
df = df.dropna()
plt.figure(figsize=(10,7))
sns.heatmap(df.isnull(), cbar = False, cmap = 'viridis')
```

```python
pd.value_counts(df['Crime type'])[:10]
```

```python
plt.figure(figsize=(14,10))
plt.title('Amount of Crimes by Crime Type')
plt.ylabel('Crime Type')
plt.xlabel('Amount of Crimes')

df.groupby([df['Crime type']]).size().sort_values(ascending=True).plot(kind='barh')

plt.show()
```

```python
Target = 'Crime type'
```

```python
print('Target: ', Target)
```

```python
# Now we are left with 14 Class as our predictive class
Classes = df['Crime type'].unique()
Classes
```

```python
df['Crime type'] = pd.factorize(df["Crime type"])[0] 
df['Crime type'].unique()
```

```python
df['Month'] = pd.factorize(df["Month"])[0]
df['Location'] = pd.factorize(df["Location"])[0]
df['LSOA code'] = pd.factorize(df["LSOA code"])[0]
df['LSOA name'] = pd.factorize(df["LSOA name"])[0]
df['Last outcome category'] = pd.factorize(df["Last outcome category"])[0]
```

```python
X_fs = df.drop(['Crime type'], axis=1)
Y_fs = df['Crime type']

#Using Pearson Correlation
plt.figure(figsize=(20,10))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
df
```

```python
cor_target = abs(cor['Crime type'])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.2]
relevant_features
```

```python
Features = ['Month','Longitude', 'Latitude', 'Location', 'LSOA code', 'LSOA name', 'Last outcome category']
```

```python
#Split dataset to Training Set & Test Set
x, y = train_test_split(df, 
                        test_size = 0.3, 
                        train_size = 0.7, 
                        random_state= 3)

x1 = x[Features]    #Features to train
x2 = x[Target]      #Target Class to train
y1 = y[Features]    #Features to test
y2 = y[Target]      #Target Class to test

print('Feature Set Used    : ', Features)
print('Target Class        : ', Target)
print('Training Set Size   : ', x.shape)
print('Test Set Size       : ', y.shape)
```

```python
rf_model = RandomForestClassifier(n_estimators=70, # Number of trees
                                  min_samples_split = 30,
                                  bootstrap = True, 
                                  max_depth = 50, 
                                  min_samples_leaf = 25)

# Model Training
rf_model.fit(X=x1,
             y=x2)

# Prediction
result = rf_model.predict(y[Features])
```

```python
ac_sc = accuracy_score(y2, result)
rc_sc = recall_score(y2, result, average="weighted")
pr_sc = precision_score(y2, result, average="weighted")
f1_sc = f1_score(y2, result, average='micro')
confusion_m = confusion_matrix(y2, result)

print("========== Random Forest Results ==========")
print("Accuracy    : ", ac_sc)
print("Recall      : ", rc_sc)
print("Precision   : ", pr_sc)
print("F1 Score    : ", f1_sc)
print("Confusion Matrix: ")
print(confusion_m)
```

```python
target_names = Classes
visualizer = ClassificationReport(rf_model, classes=target_names)
visualizer.fit(X=x1, y=x2)     # Fit the training data to the visualizer
visualizer.score(y1, y2)       # Evaluate the model on the test data

print('================= Classification Report =================')
print('')
print(classification_report(y2, result, target_names=target_names))

g = visualizer.poof()    
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)

# Model Training
knn_model.fit(X=x1,
             y=x2)

# Prediction
result = knn_model.predict(y[Features]) 
```

```python
ac_sc = accuracy_score(y2, result)
rc_sc = recall_score(y2, result, average="weighted")
pr_sc = precision_score(y2, result, average="weighted")
f1_sc = f1_score(y2, result, average='micro')
confusion_m = confusion_matrix(y2, result)

print("========== K-Nearest Neighbors Results ==========")
print("Accuracy    : ", ac_sc)
print("Recall      : ", rc_sc)
print("Precision   : ", pr_sc)
print("F1 Score    : ", f1_sc)
print("Confusion Matrix: ")
print(confusion_m)
```
